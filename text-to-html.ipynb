```python
def text_to_html(text, filename="output.html"):
    # Split the text by lines
    lines = text.split('\n')

    # Initialize an empty list to collect the HTML lines
    html_lines = []

    # Iterate over each line
    for line in lines:
        # Check if the line ends with an image or video extension
        if line.endswith('.jpg') or line.endswith('.jpeg'):
            html_lines.append(f'<img src="{line}" alt="" class="maui-image">')
        elif line.endswith('.mp4'):
            html_lines.append(f'<video src="{line}" class="maui-video" controls></video>')
        # Check if the line looks like a date
        elif any(word in line for word in ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']):
            html_lines.append(f'<h2>{line}</h2>')
        else:
            html_lines.append(f'<p>{line}</p>')

    # Join the lines into a single HTML string, wrapped in the .maui-container class
    html_content = '\n'.join(html_lines)
    html_output = f'<div class="maui-container">{html_content}</div>'

    # Write the resulting HTML to a new file   
    with open(filename, "w", encoding="utf-8") as file:
        file.write(html_output)

    return f"HTML content written to {filename}"
```
